```python
import pandas as pd

# masukin data set
file_path = '/content/camera_dataset.csv'
camera_data = pd.read_csv(file_path)


# nampilin beberapa kolom dari dataset
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
print(camera_data)
```

```python
# rata rata harga kamera
price_stats = camera_data['Price'].describe()
price_stats
```

```python
# kita ngefilter kamera yang kurang dari sama dengan 400 dan punya lensa dekat
# kenapa 400 ? karena 75% harga kamera di dataset dibawah 400, dan karena target utama kita mahasiswa yang mungkin memiliki budget terbatas. ini juga agar modal yang dikeluarkan tidak terlalu banyak
filter_kamera = camera_data[(camera_data['Price'] <= 400) & (camera_data['Macro focus range'] > 0)]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# nampiliin 5 teratas
filter_kamera.head()
```

```python
# beberapa model model kamera yang terfilter
selected_cameras = {
    'Agfa ePhoto 1280': 179.0,
    'Agfa ePhoto CL45': 179.0,
    'Canon PowerShot 350': 149.0,
    'Canon PowerShot 600': 139.0,
    'Canon PowerShot A10': 139.0
}

# hitung total modal yang diperlukan untuk membeli 50 kamera dari beberapa model tersebut.
hitungkamera = 50 // len(selected_cameras)
total_modal = sum(price * hitungkamera for price in selected_cameras.values())

# tampilkan total modal
total_modal
```
